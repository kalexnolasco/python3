# Graficas de contorno en 3D

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creamos los datos
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xx, yy = np.meshgrid(x, y, sparse=True)
z1 = np.exp(-xx**2 - yy**2)
z2 = np.exp(-(xx - 1)**2 - (yy - 1)**2)
z = (z1 - z2) * 2
```

```python
plt.figure(figsize=(15,10))

# Indicamos que usaremos proyeccion en 3d
axes=plt.axes(projection='3d')

# Hacemos la grafica de contorno pasando los datos y pidiendo 50 niveles
axes.contour3D(x,y,z,50)

plt.show()
```

```python
plt.figure(figsize=(15,10))
axes=plt.axes(projection='3d')

# Ahora indicamos que ponga solido el nivel
axes.contour3D(x,y,z,50, extend3d=True)

plt.show()
```

```python
plt.figure(figsize=(15,10))

# Indicamos que usaremos proyeccion en 3d
axes=plt.axes(projection='3d')

# Hacemos la grafica de contorno pasando los datos y pidiendo 50 niveles
axes.contour3D(x,y,z,50)

# Indicamos desde donde deseamos ver la grafica
axes.azim=-30
axes.elev=45
plt.show()
```
