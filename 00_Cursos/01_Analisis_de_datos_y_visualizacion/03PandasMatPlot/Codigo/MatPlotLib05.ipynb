# Boxplot

image.png
<BR>
El boxplot nos permite representar la distribucion de una forma rapida y obtener informacion importante. Tenemos el minimo y maximo.<BR>
El primer cuartil es el valor donde abajo de el se encuentra el 25% de las observaciones<BR>
Aparece la mediana, el 50% de las observaciones se encuentra abjao de ella<BR>
El 75% de las observaciones se encuentran abajo del tercer cuartil<BR>
Datos mas alla de los bigotes se consideran datos atipicos<BR>
La distancia intercuartil se encuentra entre el primer y tercer cuartil

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creamos los datos de ejemplo
np.random.seed(40)
vendedor1 = np.random.normal(100, 10, 200)
vendedor2 = np.random.normal(95, 35, 200)
vendedor3 = np.random.normal(90, 25, 200)
vendedor4 = np.random.normal(80, 30, 200)
datos=[vendedor1,vendedor2,vendedor3,vendedor4]
```

```python
# Hacemos el boxplot
plt.boxplot(datos)

plt.show()
```

```python
# Podemos colocar la caja con un notch
# Hacemos el boxplot
plt.boxplot(datos, notch=True)

plt.show()
```

```python
# Para colocar los outliers en un formato en particular
marcador= dict(markerfacecolor='r', marker='^')
# Hacemos el boxplot
plt.boxplot(datos, flierprops=marcador)

plt.show()
```

```python
# Si no deseamos que se muestren los outliers
plt.boxplot(datos, showfliers=False)

plt.show()
```

```python
# Para colocarlo en forma horizontal
plt.boxplot(datos, vert=False)

plt.show()
```

# Graficas de violin
Los graficos de violin son similares a los boxplot, pero tambien muestran la densidad de la probabilidad de los datos en diferentes valores.
<BR> Pueden mostrar la mediana y la media

```python
plt.violinplot(datos)
plt.show()
```

```python
# Para mostrar la media usamos showmeans=True
plt.violinplot(datos, showmeans=True)
plt.show()
```

```python
# Para mostrar la mediana usamos showmedians=True
plt.violinplot(datos, showmedians=True)
plt.show()
```

```python
plt.violinplot(datos, showmeans=True, showmedians=True)
plt.show()
```
