# Histogramas

```python
from matplotlib import pyplot as plt
import numpy as np
```

Vamos a crear un histograma con los valores que tenemos

```python
# Creamos los datos para presentar en el histograma
mu=100
sigma=10
datos = mu + sigma * np.random.randn(10000)
```

```python
# Para crear el histograma hacemos uso de hist y pasamos la fuente de informacion
plt.hist(datos)
plt.show()

# Este histograma tiene muy baja resolucion para mostrar la distribucion
```

Podemos indicar el numero de bins que nos interese


```python
# El parametro bins nos sirve para inidcar la cantidad de bins que deseamos
plt.hist(datos,bins=50)
plt.show()
```

Es posible mostrar solo el contorno

```python
# Si deseamos mostrar solo el contorno usamos histtype con el valor step
plt.hist(datos,bins=50,histtype='step')
plt.show()
```

Podemos colocar bins en valores especificos si lo necesitamos

```python
plt.hist(datos,bins=[0,70,100,110,120,130,140])
plt.show()
```

# Grafica de pie
La creacion de graficas de pie es sencilla, necesitamos los datos y las etiquetas

```python
datos=np.array([5,7,1,3,12])
etiquetas=['Juegos','Alimentos','Ropa','Calzado','Bebidas']

plt.pie(datos,labels=etiquetas)
plt.show()
```

# Graficas de dispersion
Podemos hacer tambien graficas de dispersion

```python
valoresx=[10,20,30,40,50,60]
valoresy=[12,17,18,25,31,50]

plt.scatter(x=valoresx,y=valoresy)
plt.show()
```

```python
valoresx=[10,20,30,40,50,60]
valoresy1=[12,17,18,25,31,50]
valoresy2=[45,40,38,23,18,15]

plt.scatter(x=valoresx,y=valoresy1, color='r')
plt.scatter(x=valoresx,y=valoresy2, color='b')
plt.show()
```

# Trazo de lineas entre valores

```python
valoresx=[10,20,30,40,50,60]
valoresy1=[12,17,18,25,31,50]
valoresy2=[45,40,38,23,18,15]

plt.scatter(x=valoresx,y=valoresy1, color='r')

# Plot une con lineas los puntos indicados por (valoresx,valoresy2)
plt.plot(valoresx,valoresy2, color='b')
plt.show()
```

# Graficas de contorno
Nos permiten ver los niveles de una superficie, representada en 2D

```python
# Creamos los datos en 3D que representan una superficie
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xx, yy = np.meshgrid(x, y, sparse=True)
z1 = np.exp(-xx**2 - yy**2)
z2 = np.exp(-(xx - 1)**2 - (yy - 1)**2)
z = (z1 - z2) * 2

# Para crear la grafica de contorno pasamos las listas, arreglos, series que
# contienen las coordenadas de los puntos
plt.contour(x,y,z)
plt.show()
```

```python
# Aqui colocamos un tamano a la imagen y tambien indicamos que deseamos las etiquetas de los valores

plt.figure(figsize=(15,10))

# Guardammos una referencia a la grafica de contorno en c
c=plt.contour(x,y,z)

# Inidcamos que deseamos etiquetar la grafica de contorno
# el primer parametro es la referencia a la grafica
# inline en True indica que se coloque el valor en ese contorno y no cerca de el
# fontsize es el tamano de la fuente
plt.clabel(c,inline=True,fontsize=10)

plt.show()
```

```python
# Si indicamos un color, se crea de forma monocromatica con los valores negativos punteados
# Tambien podemos indicar la cantidad de niveles deseados

plt.figure(figsize=(15,10))
c=plt.contour(x,y,z,levels=15, colors='r')
plt.clabel(c,inline=True,fontsize=10)

plt.show()
```

```python
# Usamos contourf para que la grafica se muestre rellena
# Al usar colorbar tenemos una barra que nos indica los valores por color
# https://matplotlib.org/3.3.1/tutorials/colors/colormaps.html

plt.figure(figsize=(15,10))
c=plt.contourf(x,y,z,10, cmap='plasma')

plt.colorbar(c)
plt.show()
```

```python
# Podemos colocar las etiquetas
plt.figure(figsize=(15,10))
c=plt.contour(x,y,z,10, colors='k')

plt.clabel(c,inline=True,fontsize=10)
c=plt.contourf(x,y,z,10, cmap='plasma')

plt.colorbar(c)
plt.show()
```
