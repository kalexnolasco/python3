# Visualizacion con Pandas

Veamos como usar las capacidades de visualizacion de pandas<BR>
Sus herramientas aunque sencillas nos permiten lleva a cabo analisis rapidos
y son muy faciles de usar

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib as plt
```

Ahora importamos matplotlib como inline para que jupyter nos muestre los resultados

```python
%matplotlib inline
```

Importamos los datasets<BR>
El primero es un time series

```python
df1=pd.read_csv('df1.csv', index_col=0)
print(df1)
```

```python
df2=pd.read_csv('df2.csv')
print(df2)
```

```python
df3=pd.read_csv('df3.csv')
print(df3)
```

```python
df4=pd.read_csv('linea.csv',header=None)
print(df4)
```

Veamos como podemos crear un histograma que tenga 15 divisiones

```python
# Indicamos el dataframe, la columna que nos interesa e invocamos hist
df1['A'].hist(bins=15)
```

Podemos hacer el histograma de todas las columnas

```python
# Si no indicamos la columna se crea el histograma de todas las columnas
df1.hist(bins=50)
```

Otra forma de crear un histograma es por medio de plot indicando que el tipo de grafica es un histograma

```python
# En el parametro kind de plot indicamos el tipo de grafica que nos interesa
df1['A'].plot(kind='hist', bins=15)
```

Un metodo alterno para plot

```python
df1['A'].plot.hist(bins=15)
```

Tambien podemos lleva a cabo la creacion de una grafica de area

```python
# El dataframe que nos interesa visualizar es el que hace la invocacion
df2.plot.area()
```

Podemos colocar transparencia por medio de alpha

```python
df2.plot.area(alpha=0.5)
```

Si lo necesitamos podemos crear una grafica de barras

```python
df2.plot.bar()
```

La grafica tambien puede ser colocada en pila/stack

```python
df2.plot.bar(stacked=True)
```

Si usamos line, podemos crear una grafica de linea<BR>
Es necesario pasar la coordendada (x,y) de cada punto<BR>
Como en este caso las columnas no llevar titulo, usamos su indice para indicar que columna corresponde a que coordenada

```python
df4.plot.line(x=0, y=1)
```

Podemos usar componentes de matplotlib
figsize nos sirve para indicar el tamano de la imagen
lw el ancho de la linea
ls es el estilo de la linea : -- -.  -


```python
df4.plot.line(x=0, y=1, figsize=(15,10), lw=10, ls='-')
```

La creacion de graficas de dispersion tambien es sencilla

```python
df1.plot.scatter(x='A',y='B')
```

Con el ejemplo anterior tenemos dos niveles de informacion, horizontal y vertical<BR>
Podemos agregar un tercer nivel de informacion que sea la intensidad del color, para esto usamos c e indicamos
que columna contiene esa informacion

```python
df1.plot.scatter(x='A',y='B',c='C')
```

Para mostrar las cosas por tamano en lugar de por color usamos s<BR>
