# Configuracion de la grafica

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

# Creamos dos array para trabajar con ellos
x=np.linspace(0,10,20)
y=x**2
print(x)
print(y)
```

```python
figura = plt.figure()
axes = figura.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

# Colocamos la etiqueta de cada eje
axes.set_xlabel('Dia')
axes.set_ylabel('Ganancia')

# Colocamos el titulo de la imagen
axes.set_title('Comportamiento del negocio')
```

Es posible indicar para el eje X o Y la escala logaritmica

```python
figura = plt.figure()
axes = figura.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

# El eje X usara una escala logaritmica
axes.set_yscale('log')

axes.set_xlabel('Dia')
axes.set_ylabel('Ganancia')
axes.set_title('Comportamiento del negocio')
```

Si deseamos rotular cada tick, es necesario crear un vector con el indice de tick que vamos a usar y otro vector con las
etiquetas que se colocara en cada uno

```python
# Podemos indicar las dimensiones de la figura al usar figsize
figura = plt.figure( figsize=(15,10))

axes = figura.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

# Este vector indica el indice del tick donde va la etiqueta
axes.set_xticks([0,2,4,6,8,9,10])
# Este vector tiene las etiquetas correspondientes
axes.set_xticklabels(['lunes','martes','miercoles','jueves','viernes','sabado','domingo'])

axes.set_xlabel('Dia')
axes.set_ylabel('Ganancia')
axes.set_title('Comportamiento del negocio')
```

Si los valores fueran muy grandes quiza es necesario colocar la notacion cientifica.
Necesitamos importar ticker

```python
# Importamos ticker
from matplotlib import ticker

figura = plt.figure()
axes = figura.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

# Creamos el formateador de escala
formatter = ticker.ScalarFormatter(useMathText=True)
# Indicamos que usaremos la notacion cientifica
formatter.set_scientific(True) 
# La notacion cientifica es usada para numeros menores a 1e-1 o mayores a 1e1
formatter.set_powerlimits((-1,1)) 
# Indicamos el formato al eje que nos interesa
axes.yaxis.set_major_formatter(formatter) 

axes.set_xlabel('Dia')
axes.set_ylabel('Ganancia')
axes.set_title('Comportamiento del negocio')
```

Podemos tambien colocar una rejilla

```python
figura = plt.figure()
axes = figura.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

# Indicamos que adicionamos una rejilla
# color, indica el color a usar
# alpha, el nivel de opacidad/transparencia
# linestyle la forma como se dibujan las lineas de la rejilla
# linewidth, el ancho de la rejilla
axes.grid(color='r', alpha=0.5, linestyle='--', linewidth=0.5)

axes.set_xlabel('Dia')
axes.set_ylabel('Ganancia')
axes.set_title('Comportamiento del negocio')
```

Para etiquetar el eje en dos partes, es necesario crear un 'gemelo' con twinx y ya que lo tenemos colocamos la etiqueta con
sus propiedades


```python
figura = plt.figure()
axes = figura.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, color='blue')

# Creamos un gemelo de los ejes
axes2=axes.twinx()
# Colocamos el segundo plot
axes2.plot(y, x, color='red')

# Etiquetas del eje original
axes.set_xlabel('Dia')
axes.set_ylabel('Ganancia', color='blue')
axes.set_title('Comportamiento del negocio')

# Etiqueta del segundo eje
axes2.set_ylabel('Difusion', color='red')
```

Podemos adicionar texto adentro de la grafica

```python
figura = plt.figure()
axes = figura.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

# Primero se indican las coordenadas, en termino de los valores de la grafica
# Se coloca el texto que nos interesa
# fontsize indica el tamano del texto
# color controla el color
axes.text(4,12,'Hay crecimiento', fontsize=10, color='black')

axes.set_xlabel('Dia')
axes.set_ylabel('Ganancia')
axes.set_title('Comportamiento del negocio')
```
