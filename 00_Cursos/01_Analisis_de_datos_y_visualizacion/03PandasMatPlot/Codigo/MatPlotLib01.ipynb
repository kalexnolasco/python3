# Matplotlib
Es un biblioteca de graficacion para Python<BR>
Permite tener mucho control sobre la grafica y configurarla<BR>
Se parece al sistema de graficacion de MatLab<BR>
Trabaja bien con pandas y numpy<BR>

Tenemos que usar inline para que las graficas se muestren en el notebook en algunas versiones de Jupyter

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt

# Esta linea es necesaria en algunas versiones de jupyter para mostrar
# las graficas en el mismo notebook
%matplotlib inline

import numpy as np
```

Vamos a crear unos datos de ejemplo para poder hacer las graficas

```python
x=np.linspace(0,10,20)
y=x**2
print(x)
print(y)
```

Una forma muy sencilla de crear la grafica es invocando la funcion plot y pasando la lista de x y<BR>
Cuando finalizamos de colocar todo lo necesario para esa grafica hacemos uso de show para desplegarla e indicar que esa
grafica ha finalizado

```python
plt.plot(x,y)
# Se coloca plt.show() para mostrar el plot y evitar que
# diferentes plots se encimen, es un indicador que se ha finalizado con la configuracion del plot
# cualquier configuracion extra se tomara como si fuera en otro plot
plt.show()
```

En este ejemplo tenemos dos plots en el mismo canvas/lienzo<BR>
Podemos usar label para etiquetar cada linea trazada<BR>
Si lo deseamos podemos usar loc para indicar donde colocar la descripcion:
best, upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center

```python
plt.plot(x,y, label='Linea1')
plt.plot(y,x, label='lineas2')
plt.legend(loc='center')
#plt.legend(loc='best')
plt.show()
```

Podemos adicionar etiquetas a los ejes<BR>
xlabel se usa para la etiqueta del eje horizontal<BR>
ylabel para la etiqueta del eje vertical<BR>
tile nos permite colocar un titulo a la grafica

```python
plt.plot(x,y)
plt.xlabel('Dia')
plt.ylabel('Ganancia')
plt.title('Comportamiento del Negocio')
plt.show()
```

Ahora vamos a colocar dos graficas en el mismo lienzo<BR>
Para esto imaginamos que el lienzo se divide en una reticula<BR>
Tenemos que indicar el numero de filas, y columnas de esta reticula<BR>
Y despues indicar en que seccion deseamos colocar la grafica, contandolas de forma consecutiva por fila<BR>
1 2 3<BR>
4 5 6

```python
plt.subplot(1,2,1)
plt.xlabel('Dia')
plt.ylabel('Ganancia')
plt.title('Comportamiento del Negocio')
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x)
plt.show()
```

Existe tambien una forma orientada a objetos de usar MatPlotLib<BR>
En esta creamos objetos e invocamos metodos <BR>
Primero creamos un objeto de tipo figura<BR>

```python
figura = plt.figure()
```

Adicionamos los ejes de la figura y obtenemos una referencia a los mismos<BR>
Se pasan los valores de izquierda, abajao, ancho, alto como porcentajes en el rango de 0 a 1<BR>
Con esto definimos que area del canvas usara esa grafica<BR>
Despues indicamos la grafica, las etiquetas y el titulo<BR>

```python
axes = figura.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'b')

axes.set_xlabel('Dia')
axes.set_ylabel('Ganancia')
axes.set_title('Comportamiento del negocio')
```

```python
# Para mostrarla invocamos la figura
figura
```

Ahora colocamos dos graficas<BR>
Hay que notar que usamos add_axes para indicar donde y de que tamano es cada grafica dentro del canvas

```python
fig2 = plt.figure()

axes1 = fig2.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig2.add_axes([0.4, 0.2, 0.4, 0.3]) 

# Larger Figure Axes 1
axes1.plot(y, x, 'b')
axes1.set_xlabel('Promocion')
axes1.set_ylabel('Visitantes')
axes1.set_title('Marekting')

# Insert Figure Axes 2
axes2.plot(x, y, 'r')
axes2.set_xlabel('Dia')
axes2.set_ylabel('Ganancia')
axes2.set_title('Comportamiento del Negocio');
```

Vamos a usar subplots, para esto necesitamos de la figuras y los ejes que crea la funcion subplots<BR>
Pasamos como parametros la cantidad de filas y la cantidad de columnas<BR>
Para evitar que los diferentes plots se empalmen usamos tight_layout<BR>
El valor que contendra ejes no es mas que una lista de los ejes de las figuras<BR>
Por medio de un indice indicamos con que plot vamos a trabajar<BR>

```python
fig, ejes = plt.subplots(nrows=1, ncols=2)

plt.tight_layout()

ejes[0].plot(x,y)
ejes[0].set_title("Comportamiento del negocio")
ejes[1].plot(y,x)
```

Podemos configurar el color de la linea por medio del parametro color por medio del nombre o codigo rgb en hexadecimal<BR>
Para el ancho de la linea colocamos linewidth o lw<BR>
alpha nos permite indicar la transaparencia<BR>
Podemos colocar el estilo de linea por medio de linestyle<BR>
Tambien si lo necesitamos se pueden colocar marcadores con marker, markersize nos permite colocar el tamano y markerfacecolor el color del interior. Para colocar el ancho alrededor del marcador usamos markeredgewidth y su color con markeredgecolor

```python
fig4, ejes = plt.subplots(nrows=1, ncols=2)

plt.tight_layout()

ejes[0].plot(x,y,color='red',linewidth=5,linestyle='--')
ejes[0].plot(x*2,y+10,color='blue',lw=10, alpha=0.5)
ejes[0].set_title("Comportamiento del negocio")
ejes[1].plot(y,x,color='#FFA0A1', marker='o', markersize=4, markerfacecolor='#FF1010',
             markeredgewidth=1, markeredgecolor='black')
```

Si deseamos mostrar solo un segmento de X o Y en nuestra grafica podemos hacer uso de los limites, en los que indicamos el
rango que deseamos mostrar

```python
figura = plt.figure()
axes = figura.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'b')

axes.set_xlabel('Dia')
axes.set_ylabel('Ganancia')
axes.set_title('Comportamiento del negocio')

axes.set_xlim([1,6])
axes.set_ylim([10,60])
```

Un tipo de grafica muy utilizada es la de dispersion, la creamos haciendo uso de scatter

```python
figura = plt.figure()
axes = figura.add_axes([0.1, 0.1, 0.8, 0.8])

axes.scatter(x, y)

axes.set_xlabel('Dia')
axes.set_ylabel('Ganancia')
axes.set_title('Comportamiento del negocio')
```

Es posible crear boxplots<BR>
vert indica si se dibuja vertical u horizontal



```python
data = [np.random.normal(0, std, 200) for std in range(1, 5)]
#print(data)

plt.boxplot(data,vert=True);
plt.show()
```
